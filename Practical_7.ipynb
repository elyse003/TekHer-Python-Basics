### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.


```python
x = np.linspace(0, 10, 100)
y1, y2, y3, y4 = x, x**2, x**3, x**4

fig, axes = plt.subplots(2, 2, figsize=(10, 6))

axes[0, 0].plot(x, y1, label='y = x', color='blue')
axes[0, 1].plot(x, y2, label='y = x^2', color='green')
axes[1, 0].plot(x, y3, label='y = x^3', color='red')
axes[1, 1].plot(x, y4, label='y = x^4', color='purple')

for ax in axes.flat:
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```

### Exercise 26:
Create a 3D plot of the function `z = x^2 + y^2` over a grid of `x` and `y` values.


```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Plot: z = x^2 + y^2')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=z, cmap='viridis', edgecolors='black')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot')
plt.show()
```
